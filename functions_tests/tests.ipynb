```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
phi = 1.2
beta = 0.8

A = np.linspace(0, 20, 100)
Y = phi * (A ** beta)

plt.figure(figsize=(8, 5))
plt.plot(A, Y, color='blue', linewidth=2, label='Y = 1.2 * A^0.8')

plt.title('Função 1: Produção D (Crescente e Côncava)')
plt.xlabel('A (Ativo/PL)')
plt.ylabel('Y (Produção)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

plt.show()
```

```python
alpha = 0.1

A = np.linspace(0, 20, 100)
rjt = alpha * A ** (-alpha)

plt.figure(figsize=(8, 5))
plt.plot(A, rjt, color='green', linewidth=2, label='r_jt = 0.1 * A^(-0.1)')
plt.title('Função 2: Retorno sobre Ativo (Decrescente) - U')
plt.xlabel('A (Ativo/PL)')
plt.ylabel('r_jt (Retorno sobre Ativo)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
w = 1
delta_d = 0.5
phi = 1.2
beta = 0.8

lit = (w * delta_d * phi * (A ** -(1-beta))) - 1
plt.figure(figsize=(8, 5))
plt.plot(A, lit, color='red', linewidth=2, label='L_it = w * δ_d * 1.2 * A^-(1-0.8) - 1')
plt.title('Função 3: Lucro sobre Ativo (Decrescente) - L - Downstream')
plt.xlabel('A (Ativo/PL)')
plt.ylabel('L_it (Lucro sobre Ativo)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

sigma = 0.05
theta = 0.5  

l_it_fixed = 1.0
A_zt = np.linspace(0.1, 20, 100)
r_by_A = (sigma * (A_zt ** -sigma)) + (theta * (l_it_fixed ** theta))


A_zt_fixed = 1
l_it = np.linspace(0.1, 5, 100) # Alavancagem de 0.1 até 5
r_by_l = (sigma * (A_zt_fixed ** -sigma)) + (theta * (l_it ** theta))


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(A_zt, r_by_A, color='purple', linewidth=2)
ax1.set_title(f'Variação pelo Ativo do Banco ($A_z)\n(l_it fixo={l_it_fixed})')
ax1.set_xlabel('A_zt (Ativo do Banco)')
ax1.set_ylabel('Taxa de Juros r_izt)')
ax1.grid(True, linestyle='--', alpha=0.6)

# Plot 2: r vs l_it
ax2.plot(l_it, r_by_l, color='orange', linewidth=2)
ax2.set_title(f'Variação pela Alavancagem da Firma\n(A_zt fixo={A_zt_fixed})')
ax2.set_xlabel('l_it (Alavancagem da Firma)')
ax2.set_ylabel('Taxa de Juros')
ax2.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
